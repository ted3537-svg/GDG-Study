```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)

plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

print(length)
print(weight)
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]

print(fish_data)
```

```python
fish_target = [1]*35 + [0]*14
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
```

```python
new_fish = [
    [30, 600],
    [12, 10],
    [25, 150],
    [20, 80],
    [40, 800],
    [20, 100],
    [23, 200]
]
```

```python
kn.predict(new_fish)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)

pred = kn.predict(new_fish)

for fish, p in zip(new_fish, pred):
    if p == 1:
        plt.scatter(fish[0], fish[1], marker='^', color='blue')  # 도미
    else:
        plt.scatter(fish[0], fish[1], marker='^', color='orange')  # 빙어

plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
